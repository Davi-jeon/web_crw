```python
import requests
from bs4 import BeautifulSoup
import json
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
mon = ['01', '02', '03', '04']
day = ['01', '02', '03', '04']
hour = ['01', '02', '03', '04']
data = pd.DataFrame(columns = ['CO', 'MSRADM', 'MSRDT', 'MSRRGN_CD', 'MSRRGN_NM', 'MSRSTE_NM', 'NO2', 'O3', 'PM10', 'PM24', 'PM25', 'SO2'])

row = 0

# 날짜
for m in mon:
    for d in day:
        for h in hour:
            row += 1
            res = requests.get('http://openAPI.seoul.go.kr:8088/7a664e42637461723130307248785847/json/TimeAverageCityAir/1/5/2016{}{}{}00/도심권/종로구'.format(m, d, h))
            test_data = json.loads(res.content)
            for key in test_data['TimeAverageCityAir']['row'][0].keys():
                data.loc[row, key] = test_data['TimeAverageCityAir']['row'][0][key]
            

data
```
